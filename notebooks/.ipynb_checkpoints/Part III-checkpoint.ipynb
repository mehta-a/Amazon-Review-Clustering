```python
# Confusion Matrix
```

```python
def confusion_matrix(products, clusters):
    mat = np.zeros((products, len(clusters)))
    for i in range(products):
        for j in range(len(clusters)):
            clust = clusters[j]
            mat[i][j] = len(clust)
    return mat
```

```python
confusion_matrix(k1, clusters1)
```

```python
confusion_matrix(k2, clusters2)
```
